```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# ðŸš€ Loop over multiple pages
base_url = "https://www.petitesannonces.ch/"
headers = {"User-Agent": "Mozilla/5.0"}
links = []

for page in range(1, 204):  
    url = base_url + 'r/1403?p=' + str(page)
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, "lxml")

    for a in soup.find_all('a', href=True):
        if a['href'][0:3] == '/a/':
            links.append(a['href'])
```

```python
data = []
for i,l in enumerate(links):
    url = base_url + l
    print(url)
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, "lxml")

    car_details = {}
    for row in soup.select('table.ecd tr'):
        key = row.select_one('.ecn').get_text(strip=True)
        value = row.select_one('.ecv').get_text(strip=True)
        car_details[key] = value

    description_div = soup.find('div', class_='cdk bls')
    description_text = description_div.get_text(separator='\n', strip=True)
    car_details['description'] = description_text

    for row in soup.select('div.cdk table[summary="Informations sur l\'annonce"] tr'):
        key = row.find('td', class_='ecn').get_text(strip=True)
        value = row.find_all('td')[1].get_text(strip=True)
        car_details[key] = value

    data.append(car_details)
    print('loaded:' + l + ' - #' + str(i))  
```

```python
df = pd.DataFrame(data)
print(df)

df.to_csv('petites_annonces_cars.csv')
```
